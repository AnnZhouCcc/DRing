```python
trafficname = "cluster_b"
scaledown = 100000

gurobifile = f"/home/annzhou/DRing/src/emp/datacentre/computerouting/output/spatial_dring_{trafficname}_-1_0_64_d{scaledown}"

datadict = dict()
with open(gurobifile,'r') as f:
    lines = f.readlines()
    for line in lines:
        tokens = line.split(',')
        routing = tokens[0]
        interval = int(tokens[1])
        throughput = float(tokens[2])

        if routing not in datadict: datadict[routing] = [[],[]]
        datadict[routing][0].append(interval)
        datadict[routing][1].append(throughput)


import matplotlib
import matplotlib.pyplot as plt

for routing,twolist in datadict.items():
    plt.plot(twolist[0],twolist[1],label=routing,alpha=0.8)
plt.legend()
plt.title(trafficname)
plt.xlabel("intervals")
plt.ylabel("throughput (%)")
plt.savefig(f"/home/annzhou/DRing/src/emp/datacentre/computerouting/output/spatial_{trafficname}_d{scaledown}.png", dpi=500)
plt.show()
```
